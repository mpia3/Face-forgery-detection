```python
#Caricamento modello
path_to_model = "./Esperimento 1/dataset/ResNet101V2"
model_name = "saveAfter_at_5.h5"
model = tf.keras.models.load_model(path_to_model+"/"+model_name)
IMG_SIZE = (160, 160)
```

```python
class_one = 0
class_two = 0

path_test = ["./Esperimento 1/dataset/Test/FFHQ", "./Esperimento 1/dataset/Test/StyleGAN2"]
format_file = ['png', 'png']

#for z in range(2):
for file in glob.glob(path_test[1]+"/*."+format_file[1]): # percorso dataset di una singola classe
    img = tf.keras.preprocessing.image.load_img( file, target_size=IMG_SIZE )
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)

    predictions = model.predict(img_array)
    score = predictions[0]
    

    score = tf.math.sigmoid(score)
    
    if ( 100 * (1 - score) ) > ( 100 * score ):
        class_one += 1
    else:
        class_two += 1
    
    print (class_one, class_two)
```
